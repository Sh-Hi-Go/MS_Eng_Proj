### Missing Values

##### Categorical feature

```python
features_nan=[feature for feature in car_dataset.columns if car_dataset[feature].isnull().sum()>=1 and car_dataset[feature].dtypes=='O']

def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data
```

```python
car_dataset=replace_cat_feature(car_dataset,features_nan)
print(car_dataset[features_nan].isnull().sum())
print(features_nan)
```

```python
car_dataset.head(20)
```

##### Discrete Numerical feature

```python
numerical_with_nan=[feature for feature in car_dataset.columns if car_dataset[feature].isnull().sum()>=1 and car_dataset[feature].dtypes!='O']

def replace_num_feature(dataset,features_nan):
    data=dataset.copy()
    
    for feature in features_nan:
        ## We will replace by using median since there are outliers
        median_value=dataset[feature].median()

        ## create a new feature to capture nan values
        #data[feature+'nan']=np.where(data[feature].isnull(),1,0)
        data[feature].fillna(median_value,inplace=True)
    
    return data
```

```python
car_dataset = replace_num_feature(car_dataset,numerical_with_nan)
print(car_dataset[numerical_with_nan].isnull().sum())
car_dataset.head(20)
```

### Handling rare categorical features

```python
categorical_features=[feature for feature in car_dataset.columns if car_dataset[feature].dtype=='O' and feature not in ['Body_Type','Fuel_Type','Type','_id','Make','Model','Variant','Launch_Date']]

def handle_rare_cat(dataset,features_cat):
    data=dataset.copy()
    
    for feature in features_cat:
        temp=data.groupby(feature)['Ex-Showroom_Price'].count()/len(dataset)
        temp_df=temp[temp>0.01].index
        data[feature]=np.where(data[feature].isin(temp_df),data[feature],'Rare_category')
        
    return data
```

```python
#count_company = customer_dataset.groupby('company').size()
count_company = customer_dataset['company'].value_counts()
#count_company = pd.DataFrame(count_company)
count_company.reindex()
print(type(count_company),count_company.shape)
count_company.head()
```

```python
def count_occurences(column):
    count_value = pd.DataFrame(customer_dataset[column].value_counts())
    count_value.rename(columns = {column:'occurences'}, inplace = True) # , 'index':column
    count_value = count_value.rename_axis(column)
    n = count_value['occurences'].sum()
    count_value = count_value.assign(Prob_buying = lambda x: x['occurences']/n)
    count_value.loc['Missing'] = [0, count_value['Prob_buying'].median()]
    return count_value
```

```python
count_company = count_occurences('company')
count_company.head(n=len(count_company))
```

```python
count_body = count_occurences('Body_Type')
count_body.head(n=len(count_body))
```

```python
count_fuel = count_occurences('Fuel_Type')
count_fuel.head(n=len(count_fuel))
```

```python
count_transmission = count_occurences('transmission')
count_transmission.head(n=len(count_transmission))
```

#### Competition score = probability of buying that car based on customer review

```python
def probability(comp,body,fuel,trans):
    prob = 1
    
    prob *= count_company.at[comp,'Prob_buying']
    prob *= count_body.at[body,'Prob_buying']
    prob *= count_fuel.at[fuel,'Prob_buying']
    prob *= count_transmission.at[trans,'Prob_buying']
        
    return prob
```

```python
car_dataset['Competition'] = car_dataset.apply(lambda row: probability(row['Make'],row['Body_Type'],row['Fuel_Type'],row['Type']) ,axis=1)
car_dataset.head()
```
